## 왜 의료 데이터에 '표준'이 필요한가?

> 병원 A에서 혈압을 '`SBP/DBP`'로, 병원 B에서 '`수축/이완`'으로 기록한다면 두 병원의 데이터를 합칠 때 혼란이 생김. 콘센트 규격이 나라마다 달라서 어댑터가 필요한 것과 같음

### 보건의료정보 표준화
- **HL7 FHIR:** 최신 웹 기반 의료정보 교환 표준. JSON 형식 사용
- **상호운용성(Interoperability):** 서로 다른 병원 시스템 간에 데이터를 오류 없이 주고받는 능력
- **JSON:** FHIR이 사용하는 핵심 데이터 포맷. 현재 가장 보편적인 웹 데이터 표준

### JSON이란?
- JavaScript Object Notation의 약자
- 사람도 읽기 쉽고, 컴퓨터도 읽기 쉬운 텍스트 기반 데이터 포맷
- 키(Key)와 값(Value) 쌍으로 구성

```json
{
  "patient_id": 101,
  "name": "홍길동",
  "vitals": {
    "temperature": 36.8
  }
}
```

````python
# 필요한 라이브러리 불러오기
import json
import pandas as pd

print('라이브러리 로드 완료')
````

## 가상의 FHIR 형식 환자 데이터 생성

병원 A에서 병원 B로 전송되는 가상의 환자 체온/혈압 데이터를 JSON 형식으로 만들어보기

````python
# 가상의 환자 바이탈 사인 데이터 (JSON 형식)
# 병원 A → 병원 B 전송 시뮬레이션

patient_json_str = '''
[
  {
    "patient_id": 101,
    "name": "김민준",
    "age": 45,
    "gender": "M",
    "vitals": {
      "temperature": 37.2,
      "blood_pressure": {"systolic": 130, "diastolic": 85},
      "heart_rate": 78,
      "oxygen_saturation": 98.5
    },
    "timestamp": "2026-03-01T09:00:00"
  },
  {
    "patient_id": 102,
    "name": "이서연",
    "age": 62,
    "gender": "F",
    "vitals": {
      "temperature": 38.1,
      "blood_pressure": {"systolic": 155, "diastolic": 95},
      "heart_rate": 92,
      "oxygen_saturation": 96.0
    },
    "timestamp": "2026-03-01T09:15:00"
  },
  {
    "patient_id": 103,
    "name": "박준서",
    "age": 33,
    "gender": "M",
    "vitals": {
      "temperature": 36.5,
      "blood_pressure": {"systolic": 118, "diastolic": 76},
      "heart_rate": 68,
      "oxygen_saturation": 99.2
    },
    "timestamp": "2026-03-01T09:30:00"
  },
  {
    "patient_id": 104,
    "name": "최수아",
    "age": 71,
    "gender": "F",
    "vitals": {
      "temperature": 37.8,
      "blood_pressure": {"systolic": 135, "diastolic": 88},
      "heart_rate": 85,
      "oxygen_saturation": 95.5
    },
    "timestamp": "2026-03-01T09:45:00"
  },
  {
    "patient_id": 105,
    "name": "정도윤",
    "age": 58,
    "gender": "M",
    "vitals": {
      "temperature": 36.9,
      "blood_pressure": {"systolic": 142, "diastolic": 91},
      "heart_rate": 74,
      "oxygen_saturation": 97.8
    },
    "timestamp": "2026-03-01T10:00:00"
  }
]
'''

print('JSON 데이터 문자열 생성 완료')
print(f'데이터 크기: {len(patient_json_str)} 글자')
````

## JSON 파싱 — 문자열을 Python 객체로 변환

````python
# json.loads()로 JSON 문자열을 Python 리스트/딕셔너리로 변환
patients = json.loads(patient_json_str)

print(f'파싱된 데이터 타입: {type(patients)}')
print(f'총 환자 수: {len(patients)}명')
print()
print('첫 번째 환자 정보:')
print(patients[0])
````

## 중첩된 JSON 구조에서 데이터 접근하기

> JSON 데이터는 **딕셔너리(Dictionary)** 처럼 키(key)를 사용해 값에 접근합니다.

````python
# 첫 번째 환자의 데이터 접근 예시
first_patient = patients[0]

print('=== 데이터 접근 방법 ===')
print(f'환자 ID:    {first_patient["patient_id"]}')
print(f'이름:       {first_patient["name"]}')
print(f'나이:       {first_patient["age"]}세')
print(f'체온:       {first_patient["vitals"]["temperature"]}°C')
print(f'수축기혈압: {first_patient["vitals"]["blood_pressure"]["systolic"]} mmHg')
print(f'이완기혈압: {first_patient["vitals"]["blood_pressure"]["diastolic"]} mmHg')
````

## 전체 환자 데이터를 DataFrame으로 변환

````python
# JSON의 중첩 구조를 평탄화(Flatten)하여 DataFrame으로 변환
flat_data = []
for p in patients:
    row = {
        'patient_id':       p['patient_id'],
        'name':             p['name'],
        'age':              p['age'],
        'gender':           p['gender'],
        'temperature':      p['vitals']['temperature'],
        'systolic_bp':      p['vitals']['blood_pressure']['systolic'],
        'diastolic_bp':     p['vitals']['blood_pressure']['diastolic'],
        'heart_rate':       p['vitals']['heart_rate'],
        'oxygen_sat':       p['vitals']['oxygen_saturation'],
        'timestamp':        p['timestamp']
    }
    flat_data.append(row)

df = pd.DataFrame(flat_data)
print('DataFrame 변환 완료!')
df
````

## 특정 환자 데이터 검색

````python
# =====================
# 수정해 보세요: 환자 ID로 특정 환자 정보 찾기
# =====================
target_id = 104  # 찾고 싶은 환자 ID
# =====================

# 방법 1: 리스트에서 직접 찾기
for patient in patients:
    if patient['patient_id'] == target_id:
        target = patient
        break

print(f'=== 환자 ID {target_id} 정보 ===')
print(f'이름:           {target["name"]}')
print(f'나이:           {target["age"]}세')
print(f'수축기혈압(SBP): {target["vitals"]["blood_pressure"]["systolic"]} mmHg')
print(f'이완기혈압(DBP): {target["vitals"]["blood_pressure"]["diastolic"]} mmHg')
print(f'산소포화도:      {target["vitals"]["oxygen_saturation"]}%')

# 방법 2: DataFrame 필터링
print('\n--- DataFrame으로 찾기 ---')
result = df[df['patient_id'] == target_id]
print(result)
````
